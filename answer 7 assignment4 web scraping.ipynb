```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome('C:/chromedriver/chromedriver.exe')
```

```python
driver.get('https://www.naukri.com/')
```

```python
recruiter=driver.find_element_by_xpath('//*[@id="root"]/div[1]/div/ul[1]/li[2]/a').get_attribute('href')
driver.get(recruiter)
```

```python
search=driver.find_element_by_xpath('//input[@class="sugInp"]')
search.send_keys('data science')
```

```python
button=driver.find_element_by_xpath('//button[@class="fl qsbSrch blueBtn"]')
button.click()
```

```python
hr_name=[]
designation=[]
company=[]
location=[]
skill_i_hire_for=[]

s_tag=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
for s in s_tag:
    hr_name.append(s.text)

a_tag=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
for a in a_tag:
    designation.append(a.text)
    
lastw_tag=driver.find_elements_by_xpath("//p[@class='highlightable']/a[2]/i")
for k in lastw_tag:
    company.append(k.text)
    
p_tag=driver.find_elements_by_xpath("//small[@class='ellipsis']")
for p in p_tag:
    location.append(p.text)
    
w_tag=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
for w in w_tag:
    skill_i_hire_for.append(w.text)
```

```python
print(len(hr_name))
print(len(designation))
print(len(company))
print(len(location))
print(len(skill_i_hire_for))
```

```python
dict1 = {}
dict1['HR NAME']=hr_name[0:48]
dict1['DESIGNATION']=designation[0:48]
dict1['COMPANY NAME']=company[0:48]
dict1['LOCATION']=location[0:48]
dict1['SKILLS REQUIRED']=skill_i_hire_for[0:48]
df = pd.DataFrame(dict1)
df
```
