```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome('C:/chromedriver/chromedriver.exe')
```

```python
driver.get('https://www.billboard.com/')
```

```python
hot100=driver.find_element_by_xpath("//*[@id='root']/div[2]/div[2]/nav/ul/li[3]/a").get_attribute('href')
driver.get(hot100)
```

```python
song_name=[]
artist_name=[]
last_week_rank=[]
peak_rank=[]
weeks_onboard=[]

s_tag=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for s in s_tag:
    song_name.append(s.text)

a_tag=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for a in a_tag:
    artist_name.append(a.text)
    
lastw_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for k in lastw_tag:
    last_week_rank.append(k.text)
    
p_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for p in p_tag:
    peak_rank.append(p.text)
    
w_tag=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for w in w_tag:
    weeks_onboard.append(w.text)
```

```python
print(len(song_name))
print(len(artist_name))
print(len(last_week_rank))
print(len(peak_rank))
print(len(weeks_onboard))
```

```python
dict1 = {}
dict1['SONG NAME']=song_name
dict1['ARTIST NAME']=artist_name
dict1['LAST WEEK RANK']=last_week_rank
dict1['PEAK WEEK RANK']= peak_rank
dict1['WEEKS ONBOARD']=weeks_onboard
df = pd.DataFrame(dict1)
df
```
