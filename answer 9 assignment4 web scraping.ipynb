```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome('C:/chromedriver/chromedriver.exe')
```

```python
driver.get('https://www.imdb.com/list/ls095964455/ ')
```

```python
series_name=[]
year_span=[]
genre=[]
runtime=[]
ratings=[]
votes=[]

s_tag=driver.find_elements_by_xpath("//div[@class='lister-item-content']/h3/a")
for s in s_tag:
    series_name.append(s.text)

a_tag=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for a in a_tag:
    year_span.append(a.text)
    
lastw_tag=driver.find_elements_by_xpath("//span[@class='genre']")
for k in lastw_tag:
    genre.append(k.text)
    
p_tag=driver.find_elements_by_xpath("//span[@class='runtime']")
for p in p_tag:
    runtime.append(p.text)
    
w_tag=driver.find_elements_by_xpath("//span[@class='ipl-rating-star__rating']")
for w in w_tag:
    ratings.append(w.text)
    
v_tag=driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]")
for v in v_tag:
    votes.append(v.text)
```

```python
print(len(series_name))
print(len(year_span))
print(len(genre))
print(len(runtime))
print(len(ratings))
print(len(votes))
```

```python
dict1 = {}
dict1['SERIES NAME']=series_name[0:100]
dict1['YEAR-SPAN']=year_span[0:100]
dict1['GENRE']=genre[0:100]
dict1['RUNTIME']=runtime[0:100]
dict1['RATINGS']=ratings[0:100]
dict1['VOTES']=votes[0:100]
df = pd.DataFrame(dict1)
df
```
