```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome('C:/chromedriver/chromedriver.exe')
```

```python
driver.get('http://statisticstimes.com/')
```

```python
economy=driver.find_element_by_xpath("//*[@id='top']/div[2]/div[2]")
economy.click()
```

```python
india=driver.find_element_by_xpath("//*[@id='top']/div[2]/div[2]/div/a[3]").get_attribute('href')
driver.get(india)
```

```python
gdp_statewise=driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/ul/li[1]/a").get_attribute('href')
driver.get(gdp_statewise)
```

```python
fund_dict = {}
fund_dict['state'] = []
fund_dict['gsdp_19_20'] = []
fund_dict['gsdp_18_19'] = []
fund_dict['share'] = []
fund_dict['GDP'] = []
fund_dict['gsdp_19_20 at 2011_2012']=[]
fund_dict['gsdp_18_19 at 2011_2012']=[]
```

```python
for i in range(1,20):
    r_tag = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr[1]/td[2]'.format(i))
    for d in r_tag:
        fund_dict['state'].append(d.text)

    # gsdp_19_20
    sn = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr[1]/td[3]'.format(i))
    for n in sn:
        fund_dict['gsdp_19_20'].append(n.text)
    
    # gsdp_18_19
    ind = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr[1]/td[4]'.format(i))
    for z in ind:
        fund_dict['gsdp_18_19'].append(z.text)
    
    # share
    sv = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr[1]/td[5]'.format(i))
    for s in sv:
        fund_dict['share'].append(s.text)

    # GDP
    loc = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr[1]/td[6]'.format(i))
    for l in loc:
        fund_dict['GDP'].append(l.text)
    
    # gsdp_19_20 at 2011_2012
    inv = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr[1]/td[7]'.format(i))
    for n in inv:
        fund_dict['gsdp_19_20 at 2011_2012'].append(n.text)
    
    gsdp = driver.find_elements_by_xpath('//*[@id="table_id"]/tbody/tr[1]/td[8]'.format(i))
    for m in gsdp:
        fund_dict['gsdp_18_19 at 2011_2012'].append(m.text)
    
fund_df = pd.DataFrame(fund_dict)
fund_df
```
