```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome('C:/chromedriver/chromedriver.exe')
```

```python
driver.get('https://github.com/')
```

```python
button=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
button.click()
```

```python
trending=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a").get_attribute('href')
driver.get(trending)
```

```python
title=[]
description=[]
contributors=[]
language_used=[]

h_tag=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
for i in h_tag:
    title.append(i.text)
    
d_tag=driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']")
for i in d_tag:
    description.append(i.text)
    
l_tag=driver.find_elements_by_xpath("//span[@class='d-inline-block ml-0 mr-3']")
for i in l_tag:
    language_used.append(i.text)
    
c_tag=driver.find_elements_by_xpath("//a[2][@class='muted-link d-inline-block mr-3']")
for i in c_tag:
    contributors.append(i.text)
```

```python
dict1 = {}
dict1['title']=title[0:23]
dict1['description']=description[0:23]
dict1['contributors']=contributors[0:23]
dict1['language_used']=language_used[0:23]
```

```python
df = pd.DataFrame(dict1)
df
```
